```python
def remove_paragraph_symbols(sentence: str) -> str:
    """
    <p>, </p> 기호를 없애주는 함수
    """
    sentence = re.sub('<p>|<\/p>', '', sentence)

    return sentence
```

```python
def cleansing_chinese(sentence: str) -> str:
    """
    한자를 변환하는 전처리를 하는 함수
    :param sentence: 전처리 대상 문장
    :return: 전처리 완료된 문장
    """
    # chinese character를 앞뒤로 괄호가 감싸고 있을 경우, 대부분 한글 번역임
    sentence = re.sub("\([\u2E80-\u2FD5\u3190-\u319f\u3400-\u4DBF\u4E00-\u9FCC\uF900-\uFAAD]+\)", "", sentence)
    # 다른 한자가 있다면 한글로 치환
    if re.search("[\u2E80-\u2FD5\u3190-\u319f\u3400-\u4DBF\u4E00-\u9FCC\uF900-\uFAAD]", sentence) is not None:
        sentence = hanja.translate(sentence, 'substitution')

    return sentence
```

```python
removal_list =  "‘, ’, ◇, ‘, ”,  ’, ', ·, \“, ·, △, ●,  , ■, (, ), \", >>, `, /, -,∼,=,ㆍ<,>, .,?, !,【,】, …, ◆,%,"

def cleansing_special(sentence: str) -> str:
    """
    특수문자를 전처리를 하는 함수
    :param sentence: 전처리 대상 문장
    :return: 전처리 완료된 문장
    """
    sentence = re.sub("[.,\'\"’‘”“!?]", "", sentence)
    sentence = re.sub("[^가-힣0-9a-zA-Z\\s]", " ", sentence)
    
    sentence = sentence.translate(str.maketrans(removal_list, ' '*len(removal_list)))
    
    return sentence
```

```python
def cleansing_numbers(sentence: str) -> str:
    """
    숫자를 전처리(delexicalization) 하는 함수
    :param sentence: 전처리 대상 문장
    :return: 전처리 완료된 문장
    """
    
    sentence = re.sub('[0-9]+', 'NUM', sentence)
    sentence = re.sub('NUM\s+', "NUM", sentence)
    sentence = re.sub('(NUM)+', "NUM", sentence)
    
    return sentence
```

```python
def remove_twitter(sentence: str) -> str:
    """
    트위터 주소를 없애주는 함수
    """
    sentence = re.sub('@([a-zA-Z0-9.-]+)', ' ', sentence)

    return sentence
```

```python
def remove_spaces(sentence: str) -> str:
    """
    2번 이상 연속된 공백, 문장 앞뒤 공백을 없애주는 함수
    """
    sentence = re.sub("\s+", " ", sentence)
    sentence = sentence.strip()

    return sentence
```

```python
def preprocess_sent(sentence: str = None) -> str:
    """
    모든 전처리를 수행 하는 함수
    :param sentence: 전처리 대상 문장
    :return: 전처리 완료된 문장
    """
    sent_clean = sentence
    sent_clean = remove_paragraph_symbols(sent_clean)
    sent_clean = cleansing_chinese(sent_clean)
    sent_clean = cleansing_special(sent_clean)
    sent_clean = cleansing_numbers(sent_clean)
    sent_clean = remove_twitter(sent_clean)
    sent_clean = remove_spaces(sent_clean)

    return sent_clean
```

## Build vocab, Vectorize / Split train data and valid data

```python
# 사전 훈련된 word2vec 임베딩 불러오기
from gensim.models.keyedvectors import KeyedVectors

path = "/content/drive/MyDrive/"
word2vec = KeyedVectors.load_word2vec_format(path + 'Word2Vec_300D_token.model', binary=False, encoding='utf-8')
```

```python
# 단어집 구축 및 단어별 고유한 정수 인덱스 부여 함수
def build_vocab(sentence):
    print('[Info] Build vocabulary')
    vocab_set = set(token for sent in sentence for token in sent if token in word2vec) # word2vec에 나오는 token으로 구성
    vocab = { "<pad>": 0, "<unk>": 1} # pad, unk token도 기본으로 넣어줌
    start_index = len(vocab)
    for i, token in enumerate(vocab_set):
        vocab[token] = start_index + i # 각 token에 정수 인덱스 부여
    return vocab

# 문장들을 정수 인덱스의 벡터로 변환해주는 함수 (각 문장의 각 토큰에 대해 사전에서 인덱스를 찾아 리스트로 반환)
def vectorize(vocab, sentences, one_sentence=False):
    UNK = vocab.get('<unk>') # 사전에 없는 단어 처리
    if one_sentence: # one_sentence=True: sentences는 하나의 문자열로 구성된 문장
        return [vocab.get(token, UNK) for token in sentences if token in word2vec]
    return [[vocab.get(token, UNK) for token in sent if token in word2vec] for sent in sentences] # one_Sentence=False: sentences는 문장들의 리스트로 구성됨.

vocab = build_vocab(sentence)
print('[Info] Vocabulary size=', len(vocab))
vec_sentence = vectorize(vocab, sentence)
```

```python
# train data와 valid data로 나눔
split_ratio = 0.8

n_split = int(len(sentence) * split_ratio)
trn_sentence, trn_label = vec_sentence[:n_split], label[:n_split]
val_sentence, val_label = vec_sentence[n_split:], label[n_split:]
print('[Info] Split {} data to {} for train data, {} for valid data.'.format(len(vec_sentence), len(trn_sentence), len(val_sentence)))
```

## Build dataset

```python
# Build train and valid dataset
class NewsDataset(torch.utils.data.Dataset):
    def __init__(self, vocab, sentence, label):
        self.vocab = vocab
        self.sentence = sentence
        self.label = label

    def __len__(self):
        return len(self.sentence)
    
    def __getitem__(self, idx):
        sentence = torch.LongTensor(self.sentence[idx])
        label = torch.LongTensor([self.label[idx]])
        return sentence, label

train_dataset, valid_dataset = NewsDataset(vocab, trn_sentence, trn_label), NewsDataset(vocab, val_sentence, val_label)
print('[Info] Build train and valid dataset')
```

## Build iterator

```python
#padding을 위해 max_len_for_padding 구하기

# train, valid dataset에서 가장 긴 문장의 길이를 찾는다.
train_max_len = max([len(sample[0]) for sample in train_dataset])
valid_max_len = max([len(sample[0]) for sample in valid_dataset])

max_len_for_padding = max(train_max_len, valid_max_len) # train_dataset의 최대 길이와 valid_dataset의 최대길이 중 더 긴 것을 max_len_for_padding으로 설정

print(f"train_dataset의 최대 길이: {train_max_len}")
print(f"valid_dataset의 최대 길이: {valid_max_len}")
```

```python
def padding(inputs): 
    sentence, label = list(zip(*inputs))
    sentence = torch.nn.utils.rnn.pad_sequence(sentence, batch_first=True, padding_value=0) # 문장들을 패딩하여 하나의 배치 텐서로 만듦
    # max_len_for_padding보다 길이가 짧으면 가장 뒤에 0을 추가 (모든 문장을 동일한 길이로 맞추기 위해 패딩을 추가)
    if sentence.shape[1] < max_len_for_padding:
        padding = torch.zeros((sentence.shape[0], max_len_for_padding - sentence.shape[1]), dtype=torch.long)
        sentence = torch.cat([sentence, padding], dim=1)
    label = torch.cat(label, dim=0)
    batch = [sentence, label] # 문장과 레이블을 하나의 배치로 묶어 반환
    return batch
```

```python
# train iterator와 valid iterator 만들기
BATCH_SIZE = 128

train_iterator = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=padding, drop_last = True)
valid_iterator = torch.utils.data.DataLoader(valid_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=padding, drop_last = True)
```

## Model (CNN)

```python
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, n_filters, filter_sizes, output_dim, 
                 dropout):
        
        super().__init__()
        
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        
        self.convs = nn.ModuleList([
                                    nn.Conv2d(in_channels = 1, 
                                              out_channels = n_filters, 
                                              kernel_size = (fs, embedding_dim)) 
                                    for fs in filter_sizes
                                    ])
        
        self.fc = nn.Linear(len(filter_sizes) * n_filters, output_dim)
        
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, text):

        #text = [batch size, sent len]임. 그래서 permute(1,0) 안 해줘도 됨
        
        embedded = self.embedding(text)
                
        #embedded = [batch size, sent len, emb dim]
        
        embedded = embedded.unsqueeze(1)
        
        #embedded = [batch size, 1, sent len, emb dim]
        
        conved = [F.relu(conv(embedded)).squeeze(3) for conv in self.convs]
            
        #conv_n = [batch size, n_filters, sent len - filter_sizes[n]]
        
        pooled = [F.max_pool1d(conv, conv.shape[2]).squeeze(2) for conv in conved]
        
        #pooled_n = [batch size, n_filters]
        
        cat = self.dropout(torch.cat(pooled, dim = 1))

        #cat = [batch size, n_filters * len(filter_sizes)]
            
        return self.fc(cat)
```

```python
INPUT_DIM = len(vocab)
EMBEDDING_DIM = 100
N_FILTERS = 100
FILTER_SIZES = [2,3,4]
OUTPUT_DIM = 9
DROPOUT = 0.5

model = CNN(INPUT_DIM, EMBEDDING_DIM, N_FILTERS, FILTER_SIZES, OUTPUT_DIM, DROPOUT)
```

## Pretained embeddings 적용

```python
import numpy as np

def vocab_vector(embedding, token):
    if token in embedding: # token이 word2vec에 있는 경우
        return embedding[token][:100] # token의 벡터값 반환, Embedding dimension(=300)과 맞춰줌
    return np.zeros((100,)) # Embedding dimension(=300)과 맞춰줌

def embedded_vectorize(embedding, vocab, one_sentence=False):
    return [vocab_vector(embedding, token) for token in vocab] # 리스트 만들어줌

pretrained_embeddings = embedded_vectorize(word2vec, vocab)
```

```python
# pretained_embeddings의 벡터값 하나하나 돌아주면서 tensor 만들어줌
pretrained_embeddings = torch.stack([torch.from_numpy(embedding).float() if type(embedding) == np.ndarray 
                                     else embedding for embedding in pretrained_embeddings])
```

```python
# 모델의 훈련 가능한 parameter 개수를 구하는 함수
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f'The model has {count_parameters(model):,} trainable parameters')
```

```python
import torch.optim as optim

optimizer = optim.Adam(model.parameters()) # optimizer는 Adam 사용
criterion = nn.CrossEntropyLoss() # Label 여러 개 분류 모델이기 때문에 Cross Entropy 사용
```

```python
def categorical_accuracy(preds, y):
    """
    Returns accuracy per batch, i.e. if you get 8/10 right, this returns 0.8, NOT 8
    """
    top_pred = preds.argmax(1, keepdim = True)
    correct = top_pred.eq(y.view_as(top_pred)).sum()
    acc = correct.float() / y.shape[0]
    return acc
```

```python
def train(model, iterator, optimizer, criterion):

    """
    모델을 훈련시키는 함수
    """
    
    epoch_loss = 0 # epoch에서 발생한 전체 loss, 0으로 초기화
    epoch_acc = 0 # # 에폭에서 발생한 전체 정확도, 0으로 초기화
    
    model.train() # 모델을 훈련 모드로 설정
    
    for batch in iterator: # 미니배치 단위로 데이터를 가져옴
        
        optimizer.zero_grad() # optimizer의 gradient 초기화
        
        sentences, labels = batch # batch가 [sentence, label] 모양이기 때문에 sentences와 labels를 꺼내올 수 있음
        
        predictions = model(sentences) # 모델에 문장 데이터를 입력하여 예측값 계산
        
        loss = criterion(predictions, labels) # loss(모델의 예측값과 실제 레이블 간의 손실) 계산
        
        acc = categorical_accuracy(predictions, labels) # 모델의 예측값과 실제 레이블 간의 정확도 계산
        
        loss.backward() # 손실 함수를 미분하여 gradient 계산
        
        optimizer.step() # gradient를 이용하여 가중치 파라미터 업데이트
        
        epoch_loss += loss.item() # 미니배치에서 발생한 loss를 누적하여 epoch_loss에 저장
        epoch_acc += acc.item() # 미니배치에서 발생한 정확도를 누적하여 epoch_acc에 저장
        
    return epoch_loss / len(iterator), epoch_acc / len(iterator)
```

```python
def evaluate(model, iterator, criterion):

    """
    학습된 모델을 이용해 평가 데이터셋에 대한 손실과 정확도를 계산하는 함수
    """
    
    epoch_loss = 0
    epoch_acc = 0
    
    model.eval() # 모델을 evaluation mode로 변경
    
    with torch.no_grad(): # 계산 그래프 생성을 방지하여 메모리 사용을 줄이고 평가 과정에서 학습이 일어나지 않도록 함
    
        for batch in iterator:

            sentences, labels = batch # 배치 단위로 문장과 레이블을 가져옴

            predictions = model(sentences) # model에 문장을 입력해서 나온 문장들의 레이블 예측값
            
            loss = criterion(predictions, labels) # predictions와 labels를 이용해 계산된 손실값
            
            acc = categorical_accuracy(predictions, labels) # predictions와 labels를 이용해 계산된 정확
            # 배치별로 계산된 손실과 정확도를 누적해서 학습 데이터셋 전체에 대한 손실과 정확도를 계산
            epoch_loss += loss.item()
            epoch_acc += acc.item()
        
    return epoch_loss / len(iterator), epoch_acc / len(iterator) # len(iterator)로 배치 개수로 나누어 평균 손실과 정확도를 반환
```

```python
import time

# epoch_time 얼마나 걸리는지 구함
def epoch_time(start_time, end_time):
    elapsed_time = end_time - start_time
    elapsed_mins = int(elapsed_time / 60)
    elapsed_secs = int(elapsed_time - (elapsed_mins * 60))
    return elapsed_mins, elapsed_secs
```

## Model 훈련시키기

```python
N_EPOCHS = 5 # 학습할 epoch의 수: 5로 설정

best_valid_loss = float('inf') # 현재까지 validation loss 중 최저값을 저장하기 위한 변수

for epoch in range(N_EPOCHS): # N_EPOCHS만큼 반복

    start_time = time.time() # 현재 epoch의 시작 시간
    
    train_loss, train_acc = train(model, train_iterator, optimizer, criterion) # 현재 epoch에서 학습 데이터를 사용해 모델을 학습하고 얻은 loss와 accuracy 값
    valid_loss, valid_acc = evaluate(model, valid_iterator, criterion) # 현재 epoch에서 검증 데이터를 사용해 모델을 검증하고 얻은 loss와 accuracy 값
    
    end_time = time.time() # 현재 epoch의 종료 시간

    epoch_mins, epoch_secs = epoch_time(start_time, end_time) # 현재 epoch의 학습/검증 소요 시간 (분, 초)
    
    # 현재 epoch에서 얻은 validation loss가 best_valid_loss보다 작으면 모델 저장
    if valid_loss < best_valid_loss:
        best_valid_loss = valid_loss
        torch.save(model.state_dict(), 'trained-model.pt')
    
    print(f'Epoch: {epoch+1:02} | Epoch Time: {epoch_mins}m {epoch_secs}s')
    print(f'\tTrain Loss: {train_loss:.3f} | Train Acc: {train_acc*100:.2f}%')
    print(f'\t Val. Loss: {valid_loss:.3f} |  Val. Acc: {valid_acc*100:.2f}%')
```

```python
from torch.nn.utils.rnn import pad_sequence

def predict_news(model, sentence, min_len=5, max_len=100):
    model.eval() # 모델 평가 모드로 전환
    tokenized = tokenizer(sentence) # 문장 전처리, 토큰화

    if len(tokenized) < min_len: # 문장 길이가 5보다 작으면, <pad> 추가
        tokenized += ['<pad>'] * (min_len - len(tokenized))

    UNK = vocab.get('<unk>') # '<unk>'의 고유 정수값
    vec_sentence = [vocab.get(token, UNK) for token in tokenized if token in word2vec] # word2vec에 있는 token이면 token의 고유 정수값 넣고, word2vec에 없으면 default인 UNK 넣기
    tensor = torch.LongTensor(vec_sentence) # 텐서 변환
    

    if len(tensor) > max_len: # 문장 길이가 100보다 길면, 100까지 잘라줌
        tensor = tensor[:max_len]
    else:
        pad_len = max_len - len(tensor)
        tensor = torch.nn.functional.pad(tensor, (0, pad_len), mode='constant', value=0)
    tensor = tensor.unsqueeze(0)

    # 문장을 텐서로 변환
    preds = model(tensor)
    max_preds = preds.argmax(dim=1)

    # 예측된 레이블 반환
    return max_preds.item()
```
